# PCA et Clustering

```python
fifa_dataset_pca=fifa_dataset[fifa_dataset.columns[4:]]
fifa_dataset_pca.head()
```

```python
fifa_dataset_pca=fifa_dataset_pca.fillna(fifa_dataset_pca.mean())
fifa_dataset_pca
```

```python
from sklearn.decomposition import PCA 
pca = PCA()
pca.fit(fifa_dataset_pca)

U = np.transpose(pca.components_) # vecteurs propres
C = pd.DataFrame(fifa_dataset_pca.dot(U)) # nouvelles coordonnées

I = pd.DataFrame(pca.explained_variance_ratio_)  # information véhiculée
```

```python
I
```

* On projette nos données sur les 2 premières composantes (84% de l'information)
* Remarque: Si on veut garder 95% de l'information, il faut projetter les données sur les 9 premières composantes

```python
data = C.iloc[:, [0, 1]]    
data
```

```python
from sklearn.cluster import KMeans
```

```python
inertie = []

for nb in range(1, 31):
    kmeans = KMeans(n_clusters=nb,
                    n_init=10, 
                    max_iter=300).fit(data)
    inertie.append(kmeans.inertia_)
```

fig3 plotly

```python
fig=px.line(x=list(range(1, len(inertie) + 1)), y=inertie, title="Inertie en fonction du nombre de cluster")
fig.update_xaxes(title_text='Nombre de cluster')
fig.update_yaxes(title_text='Inertie')
fig.show()
```

--> On peut donc considérer une partition en 4 clusters

```python
#Visualisation des données 
fig, ax = plt.subplots(figsize=(10, 10))

sns.scatterplot(x=0, 
                y=1, 
                data=data,
                ax=ax)

plt.show()
```

```python
#Clustering
kmeans = KMeans(n_clusters=4,
                n_init=10, 
                max_iter=300, random_state=0).fit(data)
```

```python
kmeans.n_iter_
```

```python
centers = kmeans.cluster_centers_

centers
```

```python
clusters_kmeans = kmeans.predict(data)

len(clusters_kmeans)
```

```python
kmeans.inertia_
```

```python
#couleurs 
current_palette = sns.color_palette()
colors = {i: current_palette[i] for i in range(len(centers))}
```

```python
sns.palplot(current_palette[:len(centers)]) # https://seaborn.pydata.org/tutorial/color_palettes.html
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

sns.scatterplot(x=0, 
                y=1, 
                data=data,
                hue=clusters_kmeans,
                palette=colors,
                legend=False,
                ax=ax)

sns.scatterplot(x=centers[:, 0], y=centers[:, 1], 
                hue=list(range(len(centers))),
                palette=colors,
                s=200, alpha=1,
                legend=False,
                ax=ax)



plt.show()
```

# Autres méthodes de réduction de dimension (mds et tsne)

```python
from sklearn import manifold
```

```python
def my_plot_2d_scatter(Xdata, Ydata, title="", colors=[]):
    
    fig = plt.figure(figsize=(12, 8))
    plt.scatter(Xdata, Ydata)
    plt.title(title)
    plt.show()
```

```python
df_reduc=fifa_dataset[['Overall', 'Crossing', 'Finishing',
       'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling', 'Curve',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
       'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength',
       'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'StandingTackle', 'SlidingTackle', 'GKDiving',
       'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']]
df_reduc
```

```python
n_components=2
n_neighbors=10

df_reduc= fifa_dataset_pca

mds = manifold.MDS(n_components, max_iter=100, n_init=1)
Xmds = mds.fit_transform(df_reduc)
my_plot_2d_scatter(list(Xmds[:,0]), list(Xmds[:,1]), title="MDS")

isomap = manifold.Isomap(n_neighbors, n_components)
Xiso = isomap.fit_transform(df_reduc)
my_plot_2d_scatter(Xiso[:,0], Xiso[:,1], title="Isomap")

tsne = manifold.TSNE(n_components=n_components, init='pca', random_state=0)
Xtsne = tsne.fit_transform(df_reduc)
my_plot_2d_scatter(Xtsne[:,0], Xtsne[:,1], title="t-SNE")
```

#### Faisons maintenant nos clusters 

```python
kmeans = KMeans(n_clusters=4,
                n_init=10, 
                max_iter=300, random_state=0).fit(Xtsne)

centers = kmeans.cluster_centers_

clusters_kmeans = kmeans.predict(Xtsne)

len(clusters_kmeans)

#couleurs 
current_palette = sns.color_palette()
colors = {i: current_palette[i] for i in range(len(centers))}


fig, ax = plt.subplots(figsize=(10, 10))

sns.scatterplot(x=Xtsne[:,0], 
                y=Xtsne[:,1], 
                
                hue=clusters_kmeans,
                palette=colors,
                legend=False,
                ax=ax)

sns.scatterplot(x=centers[:, 0], y=centers[:, 1], 
                hue=list(range(len(centers))),
                palette=colors,
                s=200, alpha=1,
                legend=False,
                ax=ax)

plt.title("MDS clusters")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4,
                n_init=10, 
                max_iter=300, random_state=0).fit(Xiso)

centers = kmeans.cluster_centers_

clusters_kmeans = kmeans.predict(Xiso)

len(clusters_kmeans)

#couleurs 
current_palette = sns.color_palette()
colors = {i: current_palette[i] for i in range(len(centers))}


fig, ax = plt.subplots(figsize=(10, 10))

sns.scatterplot(x=Xiso[:,0], 
                y=Xiso[:,1], 
                
                hue=clusters_kmeans,
                palette=colors,
                legend=False,
                ax=ax)

sns.scatterplot(x=centers[:, 0], y=centers[:, 1], 
                hue=list(range(len(centers))),
                palette=colors,
                s=200, alpha=1,
                legend=False,
                ax=ax)

plt.title("Isomap clusters")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4,
                n_init=10, 
                max_iter=300, random_state=0).fit(Xmds)

centers = kmeans.cluster_centers_

clusters_kmeans = kmeans.predict(Xmds)

len(clusters_kmeans)

#couleurs 
current_palette = sns.color_palette()
colors = {i: current_palette[i] for i in range(len(centers))}

fig, ax = plt.subplots(figsize=(10, 10))

sns.scatterplot(x=Xmds[:,0], 
                y=Xmds[:,1], 
                
                hue=clusters_kmeans,
                palette=colors,
                legend=False,
                ax=ax)

sns.scatterplot(x=centers[:, 0], y=centers[:, 1], 
                hue=list(range(len(centers))),
                palette=colors,
                s=200, alpha=1,
                legend=False,
                ax=ax)

plt.title("t-SNE clusters")
plt.show()
```

## Comparaison des performances des joueurs

Pour cela, on se base sur les critères suivants: "Finishing", "ShortPassing", "Strength", "Interceptions", "BallControl", "SprintSpeed", "Aggression", "SlidingTackle", 

fig4 plotly

```python
import plotly.graph_objects as go

def radar_chart(players=['Neymar Jr','L. Messi'], title="Neymar Jr Vs Leo Messi"):
    labels = ["Finishing", "ShortPassing", "Strength","Interceptions", "BallControl", "SprintSpeed", "Aggression", "SlidingTackle"]

    fig = go.Figure()
    for player in players:
        stats=np.array(fifa_dataset[fifa_dataset.Name==player][labels])[0]
        fig.add_trace(go.Scatterpolar(
              r=stats,
              theta=labels,
              fill='toself',
              name=player
        ))
    fig.update_layout(
      polar=dict(
        radialaxis=dict(
          visible=True,
          range=[0, 100]
        )),
      showlegend=True,
      title= title
    )

    fig.show()

radar_chart()
```
